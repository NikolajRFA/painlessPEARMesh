```python
from matplotlib import pyplot as plt
from tools.data_parser_2 import extract_pear_reports, df_from_pear_reports
from tools.energy_profile import get_energy_profile

pear_reports, pear_stable_time = extract_pear_reports("data/20_4_1505_1.txt")

df = df_from_pear_reports(pear_reports)

sorted_df = df.sort_values(by=['from_node', 'node_time'])

unique_node_ids = df['from_node'].unique()
```

```python
for node_id in unique_node_ids:
    period_tx, period_rx = get_energy_profile(node_id)

    filtered_df = df[df['from_node'] == node_id].copy()  # Ensures it's a separate copy
    filtered_df['node_time'] = filtered_df['node_time'] / 1000

    plt.plot(filtered_df['node_time'], filtered_df['tx_period'], marker='o', linestyle='-')
    plt.axvline(x=360000, color='red', linestyle='--', label="PEAR first run")
    plt.axvline(x=360000 + pear_stable_time, color='green', linestyle='--', label="PEAR stability reached")
    plt.axhline(y=period_tx, color='blue', linestyle='--', label="TX Energy profile")
    plt.xticks(filtered_df['node_time'], labels=filtered_df['node_time'].round(0), rotation=45)

    plt.title(f"Node {node_id}")
    plt.xlabel("Time (ms)")
    plt.ylabel("Transmissions last 60 seconds")

    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# Get unique energy profiles and assign colors
unique_profiles = {node_id: get_energy_profile(node_id)[0] for node_id in unique_node_ids}
profile_list = list(set(unique_profiles.values()))  # Unique energy profiles
color_map = plt.colormaps["tab10"]  # Use tab10 for distinct colors
profile_colors = {profile: color_map(i) for i, profile in enumerate(profile_list)}  # Map profiles to colors

plt.figure(figsize=(10, 6))

for node_id in unique_node_ids:
    period_tx, period_rx = get_energy_profile(node_id)
    filtered_df = df[df['from_node'] == node_id].copy()
    filtered_df['node_time'] = filtered_df['node_time'] / 1000

    # Assign color based on energy profile
    color = profile_colors[period_tx]

    # Remove labels from plot (so they don’t clutter the legend)
    plt.plot(filtered_df['node_time'], filtered_df['tx_period'], marker='o', linestyle='-', color=color)

# Add horizontal lines for each energy profile (these will be the only labels in the legend)
for period_tx, color in profile_colors.items():
    plt.axhline(y=period_tx, color=color, linestyle='--', linewidth=1.5, label=f"TX Energy Profile {period_tx}")

plt.axvline(x=360000, color='red', linestyle='--', label="PEAR first run")
plt.axvline(x=pear_stable_time, color='green', linestyle='--', label="PEAR stability reached")

step_size = max(len(df['node_time']) // 15, 1)
selected_ticks = (df['node_time'] / 1000)[::step_size]
plt.xticks(selected_ticks, labels=selected_ticks.round(0), rotation=45)
#plt.xticks(df['node_time'] / 1000, labels=(df['node_time'] / 1000).round(0), rotation=45)

plt.xlabel("Time (ms)")
plt.ylabel("Transmissions last 60 seconds")
plt.title("Grouped by Energy Profile TX (Distinct Colors)")
plt.legend(loc="upper right", bbox_to_anchor=(1, 1), fontsize="small")  # Legend now only shows energy profiles
plt.grid()
plt.tight_layout()
plt.show()
```

```python
temp_df = df.copy()

# Convert node_time to milliseconds if necessary
temp_df["node_time"] = temp_df["node_time"] / 1000

# Fill missing transmission values by propagating last known value
temp_df["tx_period"] = temp_df.groupby("from_node")["tx_period"].ffill()

# Aggregate total transmissions for the whole network over two-minute intervals
temp_df["time_bin"] = temp_df["node_time"] // 120000  # Convert milliseconds into 2-minute bins
df_agg = temp_df.groupby("time_bin")["tx_period"].sum().reset_index()

plt.figure(figsize=(10, 6))
# Define start and end for vertical lines
start_time = 360000
end_time = 360000 +   + 10000  # Convert to integer
# Draw vertical lines every 120000 ms starting at 360000 ms
for x in range(start_time, end_time + 1, 120000):
    plt.axvline(x=x, color="red", linestyle="--", linewidth=1, label="PEAR algorithm runs" if x == start_time else "")

plt.plot(df_agg["time_bin"] * 120000, df_agg["tx_period"], marker="o", linestyle="-", color="blue", label="Total Network Transmissions")


plt.xticks(df_agg['time_bin'] * 120000, labels=(df_agg['time_bin'] * 120000).round(0), rotation=45)

plt.xlabel("Time (ms)")
plt.ylabel("Total Messages Transmitted (2-minute window)")
plt.title("Total Messages Transmitted Over Two Minutes (Whole Network)")
plt.legend()
plt.grid()
plt.show()
```
